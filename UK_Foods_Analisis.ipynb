```python
df_sum_by_product = data.set_index('Product').sum(axis=1)

print('Suma por producto:\n', df_sum_by_product)
```

```python
df_sum_by_country = data.drop('Product', axis=1).sum()

print('Suma por país:\n', df_sum_by_country)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer estilo
sns.set(style="whitegrid")

# Gráfico de barras por producto
data.set_index('Product').plot(kind='bar', figsize=(10, 7))
plt.title('Consumo por Producto')
plt.ylabel('Consumo')
plt.xlabel('Producto')
plt.show()

# Gráfico de barras por país
data.drop('Product', axis=1).sum().plot(kind='bar', figsize=(10, 7))
plt.title('Consumo Total por País')
plt.ylabel('Consumo')
plt.xlabel('País')
plt.show()
```
